```python
import matplotlib.pyplot as plt
import random
import math
```

```python
def limcord(P):
    (x, y) = P
    return min([x, y, 1-x, 1-y])

fig, ax = plt.subplots() 

P = (random.random(), random.random())
C = limcord(P)

ax.plot((0, 0), (1, 0), 'black')
ax.plot((1, 0), (1, 1), 'black')
ax.plot((1, 1), (0, 1), 'black')
ax.plot((0, 1), (0, 0), 'black')
C = plt.Circle(P, C)
ax.add_patch(C)


#ax.plot(P[0], P[1], 'o')
fig.set_size_inches(5, 5)
plt.show()
```

```python
def dist(P0, P1):
    x0, y0 = P0
    x1, y1 = P1
    return math.sqrt((x0 - x1)**2 + (y0 - y1)**2)
cols = ['white', 'red', 'orange', 'yellow', 'green', 'blue', 'purple']
n_cols = len(cols)
def draw_pts(N):
    fig, ax = plt.subplots() 
    Rect = plt.Rectangle((0, 0), 1, 1, facecolor = 'black')
    ax.add_patch(Rect)

    L = []
    R = []
    while len(L) < N:
        x = random.random() #sigmoid(random.gauss(0, .5))
        y = random.random()
        P = (x,y)
        
                
        dist_min = [x, y, 1-x, 1-y]
        for j in range(len(L)):
            dist_min.append(abs(dist(P, L[j]) - R[j]))
        Rad = min(dist_min)
        L.append(P)
        R.append(Rad)
    
    for i in range(N):
        C = plt.Circle(L[i], R[i], facecolor = cols[i%4], alpha = .5)
        ax.add_patch(C)
    fig.set_size_inches(10, 10)
    ax.tick_params(axis='both', bottom = False, left = False, labelbottom = False, labelleft = False)
    plt.show()
    return

        
            
```

```python
draw_pts(1000)
```

```python
import numpy as np

def sigmoid(x):
    z = np.exp(-x)
    sig = 1 / (1 + z)
    return sig
```
